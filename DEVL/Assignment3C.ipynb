```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
# Fetch dataset
adult = fetch_ucirepo(id=2)

# Data (as pandas dataframes)
X = adult.data.features
y = adult.data.targets

# Display metadata and variable information (optional)
print(adult.metadata)
print(adult.variables)
```

```python
mean_age = X['age'].mean()
median_age = X['age'].median()
mode_age = X['age'].mode()[0]

print("Mean age:", mean_age)
print("Median age:", median_age)
print("Mode age:", mode_age)
```

```python
age_skewness = skew(X['age'])
print("Skewness of age:", age_skewness)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(X['age'], kde=True, bins=20, color='skyblue')
plt.axvline(mean_age, color='red', linestyle='--', label=f'Mean: {mean_age:.2f}')
plt.axvline(median_age, color='green', linestyle='--', label=f'Median: {median_age}')
plt.legend()
plt.title("Age Distribution with Density Plot")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
age_kurtosis = kurtosis(X['age'])
print("Kurtosis of age:", age_kurtosis)
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=X['age'], color='lightcoral')
plt.title("Box Plot of Age")
plt.xlabel("Age")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', data=X, palette='pastel')
plt.title("Age Distribution by Sex")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='race', y='age', data=X, palette='muted')
plt.title("Age Distribution by Race")
plt.xlabel("Race")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=X, x='age', hue='sex', fill=True)
plt.title("Density Plot of Age by Sex")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='income', y='age', data=X)
plt.title("Age Distribution by Income")
plt.xlabel("Income")
plt.ylabel("Age")
plt.show()
```
