```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
group1 = np.random.normal(loc=100, scale=10, size=30)
group2 = np.random.normal(loc=110, scale=10, size=30)
```

```python
t_stat, p_value = stats.ttest_ind(group1, group2)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)
# mean = 0, std = 1
sns.histplot(data, kde=True)
plt.title('Normal Distribution')
plt.show()
```

```python
n, p = 10, 0.5
data_binom = np.random.binomial(n, p, 1000)
sns.histplot(data_binom, kde=False)
plt.title('Binomial Distribution')
plt.show()
```

```python
lam = 5
data_poisson = np.random.poisson(lam, 1000)
sns.histplot(data_poisson, kde=False)
plt.title('Poisson Distribution')
plt.show()
```
