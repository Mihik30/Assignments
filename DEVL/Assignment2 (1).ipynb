```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print("Unique values in Fare column :")
print(df['Fare'].unique())


df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')  # Convert Fare to numeric (handles errors)
df.fillna(df['Fare'].mean(), inplace=True)

Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df.loc[df['Fare'] < lower_bound, 'Fare'] = Q1
df.loc[df['Fare'] > upper_bound, 'Fare'] = Q3
```

```python

plt.figure(figsize=(10, 6))
plt.boxplot(df['Fare'], vert=False)
plt.title('Box plot of Fare after Handling Outliers')
plt.xlabel('Fare')
plt.show()

print("Updated Fare column stats:")
print(df['Fare'].describe())
```
