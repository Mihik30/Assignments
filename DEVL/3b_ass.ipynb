```python
#1)Display all object column in Dataset
# Loop through columns of object dtype
for col in df.columns:
    if df[col].dtype == 'object':  # Check if column is of object type
        print(f"Column: {col}")
        print(df[col])  # Display the column
        print("\n")
```

```python
#2)Find all the games where the winning team scored more than 10 points.
high_scoring_games = df[df['Score'] > 10]

# Display the filtered results
high_scoring_games
```

```python
#3)Find out when the winning team scored more than 150 points and when the
#losing team scored less than 100 points.
scoring_games = df[(df['Score'] > 50)& (df['Score'] < 60)]
# Display the filtered results
scoring_games
```

```python
#4)How many games were Played in each Season?
games_per_season = df['Year'].value_counts()

# Display the result
games_per_season
```

```python
#5)What is the average winning score for each team?
average_winning_score = df.groupby('Team')['Score'].mean()

# Display the result
average_winning_score
```

```python
#6) How many games were won at home, away, and at neutral location?
location_games = df['Region Name'].value_counts()

# Display the result
location_games
```
