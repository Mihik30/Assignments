```python
# Visualizing the distribution of grades in different subjects
sns.histplot(df['Topic'], kde=True, binwidth = 30,color='blue')
plt.title('Distribution of Scores in Different Subjects')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Assuming 'df' is your DataFrame
quality_counts = df['Topic'].value_counts()

# Using Matplotlib to create a count plot
plt.figure(figsize=(8, 6))
plt.bar(quality_counts.index, quality_counts, color='pink')
plt.title('Count Plot of Dfferent Topicc')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.show()
```

```python
# Assuming 'df' is your DataFrame
plt.figure(figsize=(10, 8))

# Using Seaborn to create a swarm plot
sns.swarmplot(x="raisedhands",
              y="VisITedResources", data=df, palette='viridis')

plt.title('Swarm Plot for Grades')
plt.xlabel('SUBJECT')
plt.ylabel('Distribution')
plt.show()
```

```python
# Set the color palette
sns.set_palette("deep")

# Assuming 'df' is your DataFrame
plt.figure(figsize=(10, 6))

# Using Seaborn to create a pair plot with the specified color palette
sns.pairplot(df)

plt.suptitle('Pair Plot for DataFrame')
plt.show()
```

```python
#plotting box plot For Parents coming or Not.
sns.boxplot(x='Discussion', y='ParentAnsweringSurvey', data=df, color='Red')
plt.title('Box Plot for Parent Answering')
plt.xlabel('Range')
plt.ylabel('Boolean')
plt.show()
```

```python
#plotting box plot
sns.boxplot(x='PlaceofBirth', y='NationalITy', data=df,color='lightBlue')
plt.title('Box Plot for ')
plt.xlabel('Country')
plt.ylabel('Nationality')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'attendance' and 'grades' are columns in your DataFrame
plt.scatter(df['StudentAbsenceDays'], df['Semester'], color='blue')
plt.title('Impact of Attendance on Grades')
plt.xlabel('Attendance (%)')
plt.ylabel('Grades (%)')
plt.show()
```
